```python
face_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def face_detect(image):
    face_image=image.copy()
    face_rects = face_cascade.detectMultiScale(face_image,scaleFactor=1.2, minNeighbors=5) 
    for (x,y,w,h) in face_rects: 
        cv2.rectangle(face_image, (x,y), (x+w,y+h), (255,255,255), 5) 
    return face_image
```

```python
result=face_detect(nadia)
```

```python
plt.imshow(result, cmap="gray")
```

```python
result1=face_detect(denis)
```

```python
plt.imshow(result1, cmap="gray")
```

```python
result2=face_detect(solvay)
```

```python
plt.imshow(result2, cmap="gray")
```
